# ***Import Libraries, Save My API Key & Built My API URL***

```python
# Impoerted the requests libraries
# !pip install requests

import requests
from google.colab import userdata

# Saved my API key from Colab secrets
OPENWEATHER_API_KEY = userdata.get('OPENWEATHER_API_KEY')

# Built my base URL for the OpenWeatherMap API
BASE_URL = "http://api.openweathermap.org/data/2.5/weather"
```

# ***Extract Data***

I chose (3) cities- **London, NewYork & Tokyo**.
I extracted weather data in a structured Pandas DataFrame and ensure the column names are descriptive.

```python
import pandas as pd
import datetime

def get_weather_data(city_name):
    params = {
        'q': city_name,
        'appid': OPENWEATHER_API_KEY,
        'units': 'metric'  # You can change to 'imperial' for Fahrenheit
    }
    response = requests.get(BASE_URL, params=params)
    response.raise_for_status() # Raise an exception for HTTP errors
    return response.json()

cities = ["London", "New York", "Tokyo"]
weather_reports = []

for city in cities:
    try:
        weather_data = get_weather_data(city)

        # Extract relevant fields
        city_name = weather_data['name']
        temperature = weather_data['main']['temp']
        humidity = weather_data['main']['humidity']
        weather_condition = weather_data['weather'][0]['description'].capitalize()
        wind_speed = weather_data['wind']['speed']
        # Convert Unix timestamp to human-readable date and time
        dt_object = datetime.datetime.fromtimestamp(weather_data['dt'])
        date_time = dt_object.strftime('%Y-%m-%d %H:%M:%S')

        weather_reports.append({
            'City Name': city_name,
            'Temperature (°C)': temperature,
            'Humidity (%)': humidity,
            'Weather Condition': weather_condition,
            'Wind Speed (m/s)': wind_speed,
            'Date and Time': date_time
        })
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data for {city}: {e}")
    except KeyError as e:
        print(f"Error parsing data for {city}: Missing key {e}")

# Create a pandas DataFrame from the collected data
df_weather = pd.DataFrame(weather_reports)

# Display the DataFrame
display(df_weather)
```

# ***Transform Data***

I transformed the data and ensure all data types are appropriate for analysis. Specifically, the 'Date and Time' column, currently a string, I converted them to a datetime object.

```python
print("Original Data Types:")
display(df_weather.info())

# Convert 'Date and Time' column to datetime objects
df_weather['Date and Time'] = pd.to_datetime(df_weather['Date and Time'])

print("\nData Types After Transformation:")
display(df_weather.info())

# Display the first few rows of the transformed DataFrame to confirm
print("\nTransformed DataFrame (first 5 rows):")
display(df_weather.head())
```

# ***Load Data***

I saved the cleaned and transformed weather data for future analysis. I save it as a CSV file, which is a widely used.

```python
# Define the filename for the CSV file
csv_filename = 'weather_data.csv'

# Save the DataFrame to a CSV file
df_weather.to_csv(csv_filename, index=False)

print(f"Weather data successfully saved to {csv_filename}")
```

# ***Basic Analysis***

I proceeded in performing some basic analysis.

```python
import pandas as pd

# Perform Basic Analysis and consolidate findings for readability

print("## Basic Weather Analysis Summary\n")

# 1. Temperature Comparison
temp_comparison = df_weather[['City Name', 'Temperature (°C)']].sort_values(by='Temperature (°C)', ascending=False)
highest_temp_city = temp_comparison.iloc[0]
lowest_temp_city = temp_comparison.iloc[-1]

print("### 1. Temperature Analysis:")
print(f"- Highest Temperature: {highest_temp_city['Temperature (°C)']:.2f}°C in {highest_temp_city['City Name']}")
print(f"- Lowest Temperature: {lowest_temp_city['Temperature (°C)']:.2f}°C in {lowest_temp_city['City Name']}\n")


# 2. Humidity Comparison
humidity_comparison = df_weather[['City Name', 'Humidity (%)']].sort_values(by='Humidity (%)', ascending=False)
highest_humidity_city = humidity_comparison.iloc[0]

print("### 2. Humidity Analysis:")
print(f"- Highest Humidity: {highest_humidity_city['Humidity (%)']}% in {highest_humidity_city['City Name']}\n")


# 3. Weather Conditions Comparison
print("### 3. Weather Conditions Analysis:")
for index, row in df_weather.iterrows():
    print(f"- {row['City Name']}: {row['Weather Condition']}")

print("\n--- End of Analysis Summary ---")
```
